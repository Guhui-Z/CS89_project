```python
# use this to get stationID and correpsonding counties for a particular state
def url_to_county_station(t):
    with urllib.request.urlopen(t) as url:
        k = url.read()
    j = json.loads(k)['features']
    stationID = []
    counties = []
    for i in range(0,len(j)):
        stationID.append(j[i].get('properties').get('id'))
        counties.append(j[i].get('properties').get('name'))
    return stationID, counties
```

```python
def retrieve_station_list_json(idlist):
    for sid in idlist:
        link = 'https://api.weather.gov/stations/'+str(sid)+'observations?limit=10'
        try:
            with urllib.request.urlopen(link) as url:
                link_json = json.load(url)
            filename = link.split('stations/')[1][:-9].replace('/', '_')
            with open(filename, 'w') as f:
                json.dump(link_json, f)
            print('saved ' + filename)
            #if filename != 'us_daily.json':
            station_files.append(filename)
        except:
            print('retrieval failed' + link)
```
